### Spelling Corrector

The following notebook is an example of the project that has been implemented in AWS. I have taken the spelling corrector module from this link (https://norvig.com/spell-correct.html). This notebook is just my understanding of the workings behind the spelling corrector

The basic idea behind the spelling corrector built here here is to get all combinations of an input word by doing certain operations that are one edit distance away from the original word. These operations include: Transpose, Delete, Replace and Insert. The list of words to which these operations have been done from the root word are at edit 1 distance from the original word (will look into detail in the code). These words are then run through a corpus that has the frequency at which these words occur in the English language and the one with the max frequency is adopted which should be the intended spelling for the input wrong word

**Note:** Peter gets not only gets the custom permutations for the word being checked, he also gets the permutation of the initial permutation. What that means is that Peter checks for both words that are one edit distance from the input word as well as two edits from the root word.
